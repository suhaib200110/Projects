#### Purpose of **re** library:
- The re (regular expression) library in Python provides functions and methods for working with regular expressions, which are powerful tools for pattern matching and manipulation of text. Regular expressions are a way to describe patterns in strings, allowing you to search, extract, and manipulate text data based on those patterns. The re library allows you to perform various operations on strings using these patterns.
